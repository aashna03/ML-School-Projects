```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Mini Project - Marks Adding.csv')
df.head()
```

```python
D = {}

for i in range(0,len(df. index)):
  D.setdefault(df.iloc[i,0], []).append(df.iloc[i,1])

D
```

```python
roll = []
marks = []
for ky in D:
  roll.append(ky)
  marks.append(sum(D[ky]))

pd.DataFrame(list(zip(roll, marks)),columns =[df.columns.values[0], df.columns.values[1]]).to_csv('output.csv')
```
